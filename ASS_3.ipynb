```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

```

```python
df[['Age', 'PassengerId']] = scaler.fit_transform(df[['Age', 'PassengerId']])

```

```python
df['Survived'] = (df['Survived'] - df['Survived'].min()) / (df['Survived'].max() - df['Survived'].min())
```
